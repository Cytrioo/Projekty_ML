```python
import pandas as pd
import random

from faker import Faker
from pycaret.clustering import setup, create_model, plot_model, pull
from pycaret.datasets import get_data
```

```python
random.seed(42)
Faker.seed(42)
fake = Faker()
```

```python
#sztuczne dane to testowania
data = []

for _ in range(1000):
    data.append({
        "imie": fake.first_name(),
        "nazwisko": fake.last_name(),
        "wiek": random.randint(18, 100),
        "miasto": fake.city(),
        "wyksztalcenie": random.choice(['Wyzsze', 'Srednie', 'Podstawowe'])
    })

df = pd.DataFrame(data)
```

```python
def klastrowanie(data, ile_grup):
    setup(data=data, session_id=42)
    model = create_model("kmeans", num_clusters=ile_grup)
    return model

model = klastrowanie(df, 4)
```

```python
plot_model(model, plot='cluster')
plot_model(model, plot='elbow')
```
